```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do problema
a0 = 2.0  # Aceleração constante (m/s²)
v0 = 10.0  # Velocidade constante (m/s)
t1 = v0 / a0  # Tempo da fase 1 (s)
t2 = 2 * t1  # Tempo da fase 2 (s)
t3 = t2 + v0 / a0  # Tempo total (s)

# Funções de movimento para cada fase
def x_t(t):
    if t < t1:  # Fase 1: Aceleração constante
        return 0.5 * a0 * t**2
    elif t < t2:  # Fase 2: Velocidade constante
        x1 = 0.5 * a0 * t1**2
        return v0 * (t - t1) + x1
    else:  # Fase 3: Desaceleração constante
        x2 = v0 * (t2 - t1) + 0.5 * a0 * t1**2
        return -0.5 * a0 * (t - t2)**2 + v0 * (t - t2) + x2

def v_t(t):
    if t < t1:  # Fase 1: Aceleração constante
        return a0 * t
    elif t < t2:  # Fase 2: Velocidade constante
        return v0
    else:  # Fase 3: Desaceleração constante
        return -a0 * (t - t2) + v0

def a_t(t):
    if t < t1:  # Fase 1: Aceleração constante
        return a0
    elif t < t2:  # Fase 2: Velocidade constante
        return 0
    else:  # Fase 3: Desaceleração constante
        return -a0

# Vetor de tempo
t = np.linspace(0, t3, 1000)

# Calculando os valores para os gráficos
x = np.array([x_t(ti) for ti in t])
v = np.array([v_t(ti) for ti in t])
a = np.array([a_t(ti) for ti in t])

# Plotando os gráficos
plt.figure(figsize=(12, 8))

# Gráfico de posição x(t)
plt.subplot(3, 1, 1)
plt.plot(t, x, label='x(t)', color='blue')
plt.title('Posição x(t)')
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.grid()
plt.legend()

# Gráfico de velocidade v(t)
plt.subplot(3, 1, 2)
plt.plot(t, v, label='v(t)', color='green')
plt.title('Velocidade v(t)')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (m/s)')
plt.grid()
plt.legend()

# Gráfico de aceleração a(t)
plt.subplot(3, 1, 3)
plt.plot(t, a, label='a(t)', color='red')
plt.title('Aceleração a(t)')
plt.xlabel('Tempo (s)')
plt.ylabel('Aceleração (m/s²)')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
