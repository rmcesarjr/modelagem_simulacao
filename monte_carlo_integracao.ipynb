```python
import numpy as np
import matplotlib.pyplot as plt

# Configurações
N_values = [100, 1000, 10000, 100000]
true_integral = 1/3
estimates_integral = []
errors_integral = []

for N in N_values:
    # Gerar N amostras uniformemente em [0,1]
    x = np.random.uniform(0, 1, N)

    # Calcular a estimativa da integral
    fx = x**2
    integral_estimate = np.mean(fx)

    estimates_integral.append(integral_estimate)
    errors_integral.append(abs(integral_estimate - true_integral))

# Plotando os resultados
plt.figure(figsize=(10, 6))
plt.plot(N_values, estimates_integral, marker='o', label='Estimativa da Integral')
plt.axhline(true_integral, color='red', linestyle='--', label='Valor verdadeiro $1/3$')
plt.xscale('log')
plt.xlabel('Número de Amostras (N)')
plt.ylabel('Estimativa da Integral')
plt.title('Estimativa de $\int_0^1 x^2 dx$ por Monte Carlo')
plt.legend()
plt.grid(True)
plt.show()

# Plotando o erro
plt.figure(figsize=(10, 6))
plt.plot(N_values, errors_integral, marker='s', color='green', label='Erro absoluto')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Número de Amostras (N)')
plt.ylabel('Erro Absoluto')
plt.title('Erro na Estimativa da Integral vs Número de Amostras')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()
```
