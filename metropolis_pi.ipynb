```python
import numpy as np
import matplotlib.pyplot as plt

def metropolis_pi(n_steps=10000, epsilon=0.1):
    x, y = 0.5, 0.5  # ponto inicial
    inside = 0
    total = 0
    samples = []

    for _ in range(n_steps):
        x_new = x + np.random.uniform(-epsilon, epsilon)
        y_new = y + np.random.uniform(-epsilon, epsilon)

        # verifica se está dentro do quadrado
        if 0 <= x_new <= 1 and 0 <= y_new <= 1:
            x, y = x_new, y_new  # aceita incondicionalmente

        samples.append((x, y))
        if x**2 + y**2 <= 1:
            inside += 1
        total += 1

    pi_est = 4 * inside / total
    return pi_est, np.array(samples)

# Execução
n_steps = 500
pi, samples = metropolis_pi(n_steps, 0.1)
print(f"Estimativa de pi: {pi:.5f}")

# Visualização
plt.figure(figsize=(5,5))
plt.scatter(samples[:,0], samples[:,1], s=1, alpha=0.5)
plt.plot(samples[:,0], samples[:,1])

circle = plt.Circle((0, 0), 1, color='r', fill=False)
plt.gca().add_patch(plt.Circle((0, 0), 1, color='blue', fill=False))
plt.title("Amostras geradas pelo algoritmo de Metrópolis")
plt.xlabel("x"); plt.ylabel("y")
plt.axis("square")
plt.show()
```
