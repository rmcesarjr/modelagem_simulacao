```python
import sys
import math
import numpy as np

# 1) Inteiro vs float (exato vs aproximado)
x_int = 553
x_float = 553.089

print("1) Tipos")
print("   x_int  =", x_int,  "->", type(x_int))
print("   x_float=", x_float, "->", type(x_float))

# 2) Tamanho e precisão típica
print("\n2) Tamanho / precisão")
print("   sys.getsizeof(int)  =", sys.getsizeof(x_int), "bytes")
print("   sys.getsizeof(float)=", sys.getsizeof(x_float), "bytes")
print("   float64 epsilon     =", np.finfo(np.float64).eps)
print("   ~dígitos decimais de float64 ≈ 16")

eps = np.finfo(np.float64).eps
print(1 + eps > 1)
print(1 + eps/2 > 1)

# 3) Um exemplo clássico de erro de ponto flutuante
print("\n3) Erro de ponto flutuante")
a = 0.1 + 0.2
print("   0.1 + 0.2 =", a)
print("   (0.1 + 0.2) == 0.3 ?", a == 0.3)
print("   com 20 casas:", format(a, ".20f"))
print("   use math.isclose:", math.isclose(a, 0.3))
```
