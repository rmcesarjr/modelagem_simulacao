```python
import numpy as np
import matplotlib.pyplot as plt

# Configurações
N_values = [100, 1000, 10000, 100000, 1000000]
true_pi = np.pi
estimates_pi = []
errors_pi = []

for N in N_values:
    # Gerar N pontos aleatórios em [0,1]x[0,1]
    x = np.random.uniform(0, 1, N)
    y = np.random.uniform(0, 1, N)

    # Contar quantos pontos caem dentro do quarto de círculo
    inside = (x**2 + y**2) <= 1
    pi_estimate = 4 * np.mean(inside)

    estimates_pi.append(pi_estimate)
    errors_pi.append(abs(pi_estimate - true_pi))

# Plotando os resultados
plt.figure(figsize=(10, 6))
plt.plot(N_values, estimates_pi, marker='o', label='Estimativa de $\pi$')
plt.axhline(true_pi, color='red', linestyle='--', label='Valor verdadeiro de $\pi$')
plt.xscale('log')
plt.xlabel('Número de Amostras (N)')
plt.ylabel('Estimativa de $\pi$')
plt.title('Estimativa de $\pi$ por Monte Carlo')
plt.legend()
plt.grid(True)
plt.show()

# Plotando o erro
plt.figure(figsize=(10, 6))
plt.plot(N_values, errors_pi, marker='s', color='green', label='Erro absoluto')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Número de Amostras (N)')
plt.ylabel('Erro Absoluto')
plt.title('Erro na Estimativa de $\pi$ vs Número de Amostras')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()
```
