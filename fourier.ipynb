```python
import numpy as np
import matplotlib.pyplot as plt

def main():
  # Sinal original: soma de duas ondas senoidais
  N = 64
  t = np.arange(N)
  f1, f2 = 3, 10  # Frequências
  signal = np.sin(2 * np.pi * f1 * t / N) + 0.5 * np.sin(2 * np.pi * f2 * t / N)
  # Calcular a TDF
  X = np.fft.fft(signal)
  # Eixo de frequência
  freqs = np.fft.fftfreq(N)
  # Plotar o sinal original e seu conteúdo de frequência
  plt.figure(figsize=(10, 4))
  plt.subplot(1, 2, 1)
  plt.plot(t, signal)
  plt.title("Sinal Original")
  plt.subplot(1, 2, 2)
  plt.stem(freqs[:N//2], np.abs(X)[:N//2])
  plt.title("Espectro de Magnitude")
  plt.tight_layout()
  plt.show()

main()
```
