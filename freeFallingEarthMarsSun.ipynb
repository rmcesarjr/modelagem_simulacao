```python
import matplotlib.pyplot as plt
import numpy as np

# Constants for Earth, Mars, and the Sun
g_earth = 9.8  # gravity on Earth in m/s^2
g_mars = 3.71  # gravity on Mars in m/s^2
g_sun = 274  # gravity on the Sun in m/s^2

# Time range
t = np.linspace(0, 5, 100)

# Function for Earth
x_t_earth = -0.5 * g_earth * t**2

# Function for Mars
x_t_mars = -0.5 * g_mars * t**2

# Function for the Sun
x_t_sun = -0.5 * g_sun * t**2

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, x_t_earth, label=r'$x(t) = -\frac{1}{2}gt^2$ on Earth', color='blue')
plt.plot(t, x_t_mars, label=r'$x(t) = -\frac{1}{2}gt^2$ on Mars', color='red')
plt.plot(t, x_t_sun, label=r'$x(t) = -\frac{1}{2}gt^2$ on the Sun', color='orange')
plt.xlabel('Time (t) [s]')
plt.ylabel('Position (x) [m]')
plt.title('Comparison of Free Fall on Earth, Mars, and the Sun')
plt.legend()
plt.grid(True)
plt.show()
```
